```python
#   
feature = df[['total_rooms', 'population', 'households']]
short_df = feature.head(3)
short_df
```

```python
#  
short_df.insert(0, 'location', ['', '', ''])
short_df.head()
```

```python
#  
short_df.loc[4] = ['', 1234, 150, 320]
short_df
```

```python
# concat   
cols = ['location',	'total_rooms',	'population',	'households']
new_row = pd.DataFrame([['', 10, 20, 30]], columns=cols)

short_df = pd.concat([short_df, new_row], ignore_index=True)
short_df
```

```python
idx = 4
short_df = short_df.drop([short_df.index[idx]])
short_df
```
